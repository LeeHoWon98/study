```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./heart_2020.csv')
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

#HeartDisease - 심장병
#Smoking - 흡연
#AlcoholDrinking - 음주
#Stroke - 뇌졸증
#DiffWalking - 디프워킹
#Diabetic - 당뇨병
#Asthma - 천식
#KidneyDisease - 신장병
#SkinCancer - 피부암

df1 = pd.get_dummies(df, columns = ['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Sex', 'Diabetic', 'Asthma',
                                 'KidneyDisease', 'SkinCancer'])
df1.head()
```

```python
df1['AgeCategory'].value_counts()
```

```python
def get_categorical(age):
    cat = ''
    if age == '18-24': cat = 18
    elif age == '25-29': cat = 25
    elif age == '30-34': cat = 30
    elif age == '35-39': cat = 35
    elif age == '40-44': cat = 40
    elif age == '45-49': cat = 45
    elif age == '50-54': cat = 50
    elif age == '55-59': cat = 55
    elif age == '60-64': cat = 60
    elif age == '65-69': cat = 65
    elif age == '70-74': cat = 70
    elif age == '75-79': cat = 75
    elif age == '80 or older': cat = 80
    
    return cat

df1['Age'] = df1['AgeCategory'].apply(lambda x : get_categorical(x))

df1.head()
```

```python
df1 = df1.drop('AgeCategory', axis = 1)
df1 = df1.drop('Race', axis = 1)
df1 = df1.drop('PhysicalActivity', axis = 1)

df1.head()
```

```python
df1 = df1.drop('GenHealth', axis = 1)
```

```python
df1 = df1.drop('Sex_Female', axis = 1)
df1 = df1.drop('Sex_Male', axis = 1)
```

```python
#로지스틱 회귀를 사용해볼것이므로 StandardScaler를 사용하여 표준화를 진행하겠습니다.
from sklearn.preprocessing import StandardScaler

X_num = df1[['BMI', 'PhysicalHealth','MentalHealth', 'SleepTime','Age']]
X_cat = df1[['Smoking_No' , 'Smoking_Yes' , 'Diabetic_No, borderline diabetes', 'Diabetic_Yes', 'Diabetic_Yes (during pregnancy)',
             'Asthma_No', 'Asthma_Yes', 'KidneyDisease_No', 'KidneyDisease_Yes', 'SkinCancer_No', 'SkinCancer_Yes',
             'AlcoholDrinking_Yes', 'AlcoholDrinking_No', 'Stroke_Yes', 'Stroke_No', 'DiffWalking_Yes', 'DiffWalking_No']]
y = df1[['HeartDisease_Yes']]
scaler = StandardScaler() 
scaler.fit(X_num)
X_scaled = scaler.transform(X_num)
X_scaled = pd.DataFrame(data=X_scaled, index=X_num.index, columns=X_num.columns)
X = pd.concat([X_scaled, X_cat], axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
pred = model_lr.predict(X_test)
print(classification_report(y_test, pred))
```
